# Quicksort


## Algorithm



- Pick the last item in an array
- First pointer that just starts at the beginning
- A second that just walks the array
- Everytime we find one that's smaller than our pivot, we put it in that first position
- Increment our runner
- So, we're putting all the ones that are smaller than, or equal to our pivot on one side of an array,
- Then our pivot, all the ones that are bigger than it on the other side. 
- This starts to look like sort
