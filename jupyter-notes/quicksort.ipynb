# Quicksort


## Algorithm



- Pick the last item in an array
- First pointer that just starts at the beginning
- A second that just walks the array
- Everytime we find one that's smaller than our pivot, we put it in that first position
- Increment our runner
- So, we're putting all the ones that are smaller than, or equal to our pivot on one side of an array,
- Then our pivot, all the ones that are bigger than it on the other side. 
- This starts to look like sort




### Example on a 32 length array







- Everything left of the pivot needs to be lss than or equal to.
- Everything right of it needs to be greater than it.



## Runtime:

### Best case scenario

n/2^k = 1 

O(N log N)

### Worst case scenario

Reverse a sorted array. -> O(N^2)


## Recursive Nature: Quicksort is a recursive algorithm. 
What condition is essential to include in the quicksort function to prevent it from entering an infinite recursion?

The base case is checking the indices of the segment being sorted. if the start index is greater than or equal to the end index, the recursion stops.
